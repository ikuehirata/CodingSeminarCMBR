# Lesson 2 - Exercise 0
Simulate free fall: 
A ball is dropped from a height of 57 m.

1. Draw a velocity-time graph, provided the object follow Newton's motion equation
$$ m a = m g $$
where $ m $ is the weight of the object, a the acceleration, 
$ g $ the gravitational acceleration 9.8 m/s<sup>2</sup>, given no other disturbance.
2. Draw height-time graphs;
    1. Analytically using
    $$ h_A(t) = h_{A\mathrm{init}} - \frac{1}{2} g t^2 $$
    2. Numerically using
    $$ h_B(t+ \Delta t) = h_B(t) + v(t) \Delta t $$

```python
%matplotlib inline
import numpy as np
dt = 0.01 # delta t
t = np.arange(0, 4, dt) # time list
a = 9.8 # acceleration
v = t * a # velocity

import matplotlib.pyplot as plt
plt.plot(t, v)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.show()
```

```python
hinit = 57
# mode A: hA(t) = h_init - 1/2 g*t^2
ha = hinit - 1/2 * a * np.power(t, 2)

# mode B: hB(t + dt) = hB(t) - v(t)dt
hb = np.empty_like(t) # creates same size array as t
hb[0] = hinit # initial height
dt = t[1] - t[0] # dt as range function step
for i in range(len(t)-1): # we want to process hb[1:-1]
    hb[i+1] = hb[i] - v[i] * dt

plt.ylim(0, 60)
plt.plot(t, ha, label="Analytical")
plt.plot(t, hb, label="Numerical")
plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.show()
```
