## One dimentional random walk
Simulate 5-step one dimentional random walk in an array and plot in $x-t$ graph:

- a dot is positioned at $x = 0$ at $t = 0$

- every second the dot moves either left or right at 50% probability *i.e.*

$$
    x(t+1) =
        \begin{cases}
            x(t)+1 \\
            x(t)-1 \\
        \end{cases}
$$

Increase the number of step to 20 when you have successfully done 5 steps.

## Two dimentional random walk
Simulate the same in two dimentional space. Plot the movement of the dot in the space.

The position of the dot $\boldsymbol{r}$ is described as:

$$ \boldsymbol{r}(t+1) = 
 \begin{cases}
    \boldsymbol{r}(t) + \left(
        \begin{array}{c}
            1\\0
        \end{array}
    \right)\\
    \boldsymbol{r}(t) + \left(
        \begin{array}{c}
            -1\\0
        \end{array}
    \right)\\
    \boldsymbol{r}(t) + \left(
        \begin{array}{c}
            0\\1
        \end{array}
    \right)\\
    \boldsymbol{r}(t) + \left(
        \begin{array}{c}
            0\\-1
        \end{array}
    \right)
 \end{cases}
$$

## Model answer
### 1-D random walk

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

length = 20
t = np.arange(length) # time
x = np.zeros(length) # position

# random walk process
for i in t[1:]: # do the movement from 1
    r = np.random.rand() # random number
    if r < 0.5:
        x[i] = x[i-1] + 1
    else:
        x[i] = x[i-1] - 1

plt.plot(t, x)
```

### 1-D random walk: shorter version
`numpy.random.choice` gives you a random choice from the list in the option.

```python
import matplotlib.pyplot as plt
import numpy as np

length = 20
t = np.arange(length) # time
x = np.zeros(length) # position
r = np.random.choice([-1, 1], length-1) # random array of -1 and +1

# random walk process
for i in t[1:]: # do the movement from 1
    x[i] = x[i-1] + r[i-1]

plt.plot(t, x)
```

### 2-D random walk
Creating 2-D `array` helps the trouble a lot.

```python
length = 20
t = np.arange(length) # time
p = np.zeros((length, 2)) # position
c = np.random.choice([0, 1], length-1) # random array of 0 or 1 to determine movement direction
r = np.random.choice([-1, 1], length-1) # random array of -1 and +1

# random walk process
for i in t[1:]: # do the movement from 1
    p[i] = p[i-1]
    p[i,c[i-1]] = p[i-1,c[i-1]] + r[i-1] # choose "column" by c

plt.plot(p[:,0], p[:,1])
for i in t:
    plt.annotate(t[i], p[i], xycoords="data") # adds number to the plotting area
```

### 3-D random walk
You can expand $2$ in above to a variable named `dimension` and enhance the dimension of your wish.

```python
length = 20
dim = 3 # random walk dimension
t = np.arange(length) # time
p = np.zeros((length, dim)) # position
c = np.random.choice(list(range(dim)), length-1) # random array to determine movement direction
r = np.random.choice([-1, 1], length-1) # random array of -1 and +1

# random walk process
for i in t[1:]: # do the movement from 1
    p[i] = p[i-1]
    p[i, c[i-1]] = p[i-1, c[i-1]] + r[i-1]

# plotting
from mpl_toolkits.mplot3d import Axes3D # this is necessary to plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(p[:,0], p[:,1], p[:,2])
for i in t:
    ax.text(p[i,0], p[i,1], p[i,2], t[i]) # adds number to the plotting area
```
